```python
import pandas as pd
import numpy as np
import scipy.io
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler 
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mat = scipy.io.loadmat(r'C:\Users\MOBS\Desktop\datasetSUA5.mat')
data = pd.DataFrame(np.hstack((mat['X'], mat['Y'])))
data[pd.isnull(data)]=0
```

```python
Y=data[162];
data.pop(162);
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, Y, test_size=0.3,random_state=12)
```

```python
nn=MLPClassifier()
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
X_train=scaler.transform(X_train)
X_test = scaler.transform(X_test)  
```

```python
nn.fit(X_train,y_train)
```

```python
plt.plot(nn.loss_curve_)
```

```python
nn=MLPClassifier(hidden_layer_sizes =(250,100,50),max_iter=1000)
```

```python
nn.fit(X_train,y_train)
```

```python
plt.plot(nn.loss_curve_)
nn.score(X_test,y_test)
```

```python
nn=MLPClassifier(hidden_layer_sizes =(300,300,300),max_iter=1000, tol=10**-6)
nn.fit(X_train,y_train)
plt.plot(nn.loss_curve_)
nn.score(X_test,y_test)
```

```python
nn=MLPClassifier(hidden_layer_sizes =(250,100,100,20),max_iter=1000, tol=10**-6,alpha=0.000001)
nn.fit(X_train,y_train)
plt.plot(nn.loss_curve_)
nn.score(X_test,y_test)
```
